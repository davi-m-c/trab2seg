```python
# Converte uma lista de bytes em uma matriz 4x4
def bytes_to_block(bytes_list):
    matrix = [[0] * 4 for _ in range(4)]
    for i in range(4):
        for j in range(4):
            matrix[j][i] = bytes_list[i * 4 + j]
    return matrix
```

```python
# Função para converter uma matriz de estado 4x4 em uma lista de bytes
def block_to_bytes(matrix):
    return [matrix[i][j] for j in range(4) for i in range(4)]

bloco = [[162, 42, 138, 106],
[234, 170, 168, 155],
[172, 234, 42, 168],
[170, 186, 234, 170]]

block_to_bytes(bloco)
```

```python
def galois_multiplication(a,b):
    if b == 1:
        return a
    else:
        # Certifique-se de que a e b estejam dentro do intervalo correto usando operadores de módulo
        soma = hex((L_TABLE[a] + L_TABLE[b]))
        soma = int(soma,16)
        while soma>0xff:
            soma-=0xff
        return hex(E_TABLE[soma])

print(galois_multiplication(234,42))
```

```python
# Função para multiplicar uma coluna da matriz de estado com a matriz de multiplicação
def mixcolumns(matrix):
    # Matriz fixa de multiplicação
    multiplication_matrix = [
        [2, 3, 1, 1],
        [1, 2, 3, 1],
        [1, 1, 2, 3],
        [3, 1, 1, 2]
    ]
    
    # Lista para armazenar o resultado
    result = [[0] * 4 for _ in range(4)]
    for i in range(4):
        for j in range(4):
            # Realiza as multiplicação e XOR para calcular cada valor da coluna
            x1 = galois_multiplication(matrix[0][j],multiplication_matrix[i][0])
            if isinstance(x1, int):
                pass
            else:
                x1 = int(x1, 16)  

            x2 = galois_multiplication(matrix[1][j],multiplication_matrix[i][1])
            if isinstance(x2, int):
                pass
            else:
                x2 = int(x2, 16) 

            x3 = galois_multiplication(matrix[2][j],multiplication_matrix[i][2])
            if isinstance(x3, int):
                pass
            else:
                x3 = int(x3, 16) 

            x4 = galois_multiplication(matrix[3][j],multiplication_matrix[i][3])
            if isinstance(x4, int):
                pass
            else:
                x4 = int(x4, 16)

            result[i][j] = (
                (x1) ^
                (x2) ^
                (x3) ^
                (x4)
            ) % 256  # Garante que o resultado seja um byte (8 bits)
    
    return result

# Exemplo de lista de bytes
bytes_list = [162, 234, 172, 170, 42, 170, 234, 186, 138, 168, 42, 234, 106, 155, 168, 170]

# Converte a lista de bytes em uma matriz de estado 4x4
matriz_estado = bytes_to_block(bytes_list)
print("Matriz de Estado antes da operação MixColumns:")
for row in matriz_estado:
    print(row)

# Aplica a operação MixColumns na matriz de estado
matriz_estado_mixed = mixcolumns(matriz_estado)
print("\nMatriz de Estado após a operação MixColumns:")
for row in matriz_estado_mixed:
    print(row)

# Converte a matriz de estado de volta para uma lista de bytes
bytes_list_mixed = block_to_bytes(matriz_estado_mixed)
print("\nLista de bytes após a operação MixColumns:")
print(bytes_list_mixed)
```

```python
def mix_columns_inverse(state):
    # Matriz inversa para MixColumns
    inverse_matrix = [
        [0x0E, 0x0B, 0x0D, 0x09],
        [0x09, 0x0E, 0x0B, 0x0D],
        [0x0D, 0x09, 0x0E, 0x0B],
        [0x0B, 0x0D, 0x09, 0x0E]
    ]

    # Lista para armazenar o resultado
    result = [[0] * 4 for _ in range(4)]

    for i in range(4):
        for j in range(4):
            x1 = galois_multiplication(state[0][j],inverse_matrix[i][0])
            if isinstance(x1, int):
                pass
            else:
                x1 = int(x1, 16)  

            x2 = galois_multiplication(state[1][j],inverse_matrix[i][1])
            if isinstance(x2, int):
                pass
            else:
                x2 = int(x2, 16) 

            x3 = galois_multiplication(state[2][j],inverse_matrix[i][2])
            if isinstance(x3, int):
                pass
            else:
                x3 = int(x3, 16) 

            x4 = galois_multiplication(state[3][j],inverse_matrix[i][3])
            if isinstance(x4, int):
                pass
            else:
                x4 = int(x4, 16)
            # Realiza as multiplicação e XOR para calcular cada valor da coluna
            result[i][j] = (
                (x1) ^
                (x2) ^
                (x3) ^
                (x4)
            ) % 256  # Garante que o resultado seja um byte (8 bits)

    return result

# Exemplo de matriz de estado após ShiftRows e SubBytes
state = [124, 40, 238, 244, 225, 250, 154, 81, 44, 85, 83, 200, 96, 14, 95, 194]

state= bytes_to_block(state)

print("Matriz de Estado antes de MixColumnsInverse:")
for row in state:
    print(row)

# Aplica MixColumnsInverse na matriz de estado
state_mixed_inverse = mix_columns_inverse(state)

print("\nMatriz de Estado após MixColumnsInverse:")
for row in state_mixed_inverse:
    print(row)
```

```python
def deshift_rows(matrix):
    for i in range(1, 4):
        matrix[i] = matrix[i][-i:] + matrix[i][:-i]
    return matrix

# Sua lista de bytes 
bytes_list = [162, 170, 42, 170, 42, 168, 168, 170, 138, 155, 172, 186, 106, 234, 234, 234]
# Converte a lista de bytes para uma matriz 4x4
matriz_estado = bytes_to_block(bytes_list)
print("Matriz de Estado:")
for row in matriz_estado:
    print(row)

# Desfaz o ShiftRows na matriz de estado
matriz_estado_deshifted = deshift_rows(matriz_estado)
print("\nMatriz de Estado após o DeshiftRows:")
for row in matriz_estado_deshifted:
    print(row)

# Converte a matriz de estado deslocada de volta para uma lista de bytes

# Cria bytes_list_shifted a partir de matriz_estado_deshifted
bytes_list_shifted = [matriz_estado_deshifted[j][i] for i in range(4) for j in range(4)]

print("\nbytes_list_shifted:")
print(bytes_list_shifted)
```
